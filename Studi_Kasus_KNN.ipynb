Untuk praktik ini menggunakan dataset klasik dari Kaggle yaitu Social Network Ads (memprediksi user membeli barang atau tidak berdasarkan Age dan Estimated Salary).


```python
# -- coding: utf-8 --
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
```

```python
url_kaggle_dataset = 'https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv'
df = pd.read_csv(url_kaggle_dataset)
```

```python
# Menampilkan 5 baris pertama data
print("--- Cuplikan Data Diabetes (Kaggle) ---")
print(df[['Glucose', 'BMI', 'Age', 'Outcome']].head())
print("\n")
```

```python
# Data Preprocessing sederhana:
# Hapus data yang kosong (NaN) jika ada agar KNN berjalan lancar
df = df.dropna()
```

```python
# Fitur & target (Kita ambil 3 fitur penting saja)
X = df[['Glucose', 'BMI', 'Age']] # Fitur: Glukosa, Indeks Massa Tubuh, Umur
y = df['Outcome']                 # Target: 1 (Diabetes), 0 (Tidak Diabetes)
```

```python
# Normalisasi (Sangat penting karena skala Glukosa bisa sampai 200, sedangkan BMI dan Umur puluhan)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# Model KNN (Kita coba tetangga k=5)
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
# Prediksi
y_pred = model.predict(X_test)
```

```python
# Evaluasi
print("--- Evaluasi KNN Diabetes ---")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))
```

```python
# ANALISIS
print("\n=======================================================")
print("ANALISIS KNN (DATASET KAGGLE DIABETES):")
print("1. Model KNN digunakan untuk memprediksi potensi diabetes (Outcome 1) atau tidak (0) berdasarkan jarak kemiripan fitur kesehatan pasien (Glukosa, BMI, Umur).")
print("2. Normalisasi (StandardScaler) sangat esensial di sini agar fitur Glukosa yang rentang angkanya besar tidak mendominasi fitur lain dalam perhitungan jarak Euclidean KNN.")
print("3. Dengan pendekatan ini, mesin mengklasifikasikan status pasien dengan melihat mayoritas status dari 5 pasien (k=5) yang rekam medisnya paling mirip.")
print("=======================================================")
```
